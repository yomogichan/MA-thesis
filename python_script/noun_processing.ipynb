```python
# Create a list of capital letter, including umlauts.
cap_letter = []

for l in string.ascii_uppercase:    
    cap_letter.append(l)

for l in ['Ö','Ä','Ü']:
    cap_letter.append(l)

print(cap_letter)
```

```python
# Create the list of noun
nouns_temp = pd.DataFrame()

for l in cap_letter:
    appended_data = df[df.Word.str.startswith(l)]
    data_temp = pd.concat([nouns_temp,appended_data],ignore_index=True)
    nouns_temp = data_temp

nouns_temp  
# nouns_temp.to_csv('noun_temp.csv',header=True,index=False,encoding="latin-1")
```

```python
noun_aoa['Frequency'] = noun_aoa['Word'].map(freq_dict)
noun_aoa
```

```python
# Normalize frequency

def normalize(df,feature_name):
    result = df.copy()
    
    max_value = df[feature_name].max()
    min_value = df[feature_name].min()
    result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
final_noun_aoa = normalize(noun_aoa,'Frequency')
final_noun_aoa
# final_noun_aoa.to_csv("noun_aoa_freq.csv",header = True, index = False, encoding = "utf8")
final_noun_aoa.to_excel("noun_aoa_freq.xls",header = True, index = False, encoding = "utf8")
```
